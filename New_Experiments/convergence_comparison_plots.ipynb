```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
pqp example
'''

true_solution = np.array([0.5,-0.5],dtype = np.float32)


DR_sol = np.load('pqpDRIdentity.npy')
ADMM_sol = np.load('pqpADMM.npy')
ADMMH_sol = np.load('pqpADMMHeur.npy')
DR_trained_sol = np.load('pqpDR.npy')

DR_cnv = np.sum( (DR_sol - true_solution)**2,1)
DRT_cnv = np.sum( (DR_trained_sol - true_solution)**2,1)
ADMM_cnv = np.sum( (ADMM_sol - true_solution)**2,1)
ADMMH_cnv = np.sum( (ADMMH_sol - true_solution)**2,1)

iters = np.arange(0,len(DR_cnv),1)
plt.plot(iters,np.log10(DR_cnv))
plt.plot(iters,np.log10(DRT_cnv))
plt.plot(iters,np.log10(ADMM_cnv))
plt.plot(iters,np.log10(ADMMH_cnv))
plt.xlabel('Iteration')
plt.ylabel('log(MSE)')
plt.legend(['DR','DR Trained','ADMM','ADMMH'])
```
